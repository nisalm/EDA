```python
# Import necessary libraries
import pandas as pd
import seaborn as sns

# Load Titanic dataset
titanic_df = sns.load_dataset('titanic')

# Display the first 5 rows of the dataset
print(titanic_df.head())

# Get the structure of the data
print(titanic_df.info())
```

```python
# Summary statistics of numerical columns
print(titanic_df.describe())

# Count of unique values for categorical features
print(titanic_df['pclass'].value_counts())
print(titanic_df['sex'].value_counts())

# Explore survival rate by passenger class
print(titanic_df.groupby('pclass')['survived'].mean())

# Explore survival rate by gender
print(titanic_df.groupby('sex')['survived'].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Distribution of Age
plt.figure(figsize=(8,6))
sns.histplot(titanic_df['age'].dropna(), kde=True)
plt.title('Distribution of Passenger Ages')
plt.show()

# Boxplot of fares by passenger class
plt.figure(figsize=(8,6))
sns.boxplot(x='pclass', y='fare', data=titanic_df)
plt.title('Fare Prices by Passenger Class')
plt.show()

# Bar plot of survival rate by gender
plt.figure(figsize=(8,6))
sns.barplot(x='sex', y='survived', data=titanic_df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
def display_univariate_counts(data, column):
    """
    Displays the count of each category in a given column of a DataFrame.

    Parameters:
    data (pd.DataFrame): The input DataFrame.
    column (str): The name of the column to display counts for.

    Returns:
    None: Displays a bar plot of category counts.
    """
    # Get the counts for each category
    category_counts = data[column].value_counts()

    # Plot the category counts
    plt.figure(figsize=(8, 6))
    sns.barplot(x=category_counts.index, y=category_counts.values)
    plt.title(f'Count of Each Category in {column}')
    plt.xlabel(f'{column}')
    plt.ylabel('Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Example usage with the Titanic dataset
display_univariate_counts(titanic_df, 'sex')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def display_numerical_distribution(data, column, bins=20, kde=True):
    """
    Displays the distribution of a numerical variable as a histogram.

    Parameters:
    data (pd.DataFrame): The input DataFrame.
    column (str): The name of the numerical column to display distribution for.
    bins (int): Number of bins for the histogram. Default is 20.
    kde (bool): Whether to overlay a kernel density estimate (KDE). Default is True.

    Returns:
    None: Displays a histogram of the numerical variable's distribution.
    """
    # Plot the histogram
    plt.figure(figsize=(8, 6))
    sns.histplot(data[column].dropna(), bins=bins, kde=kde)
    plt.title(f'Distribution of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

# Example usage with the Titanic dataset
display_numerical_distribution(titanic_df, 'age')
```

```python
# Example usage with the Titanic dataset
display_numerical_distribution(titanic_df, 'parch')
```

```python
import matplotlib.pyplot as plt

# Select only the numerical columns from the Titanic dataset
numeric_columns = titanic_df.select_dtypes(include=['float64', 'int64'])

# Calculate the correlation matrix
corr_matrix = numeric_columns.corr()

# Display the correlation values for 'survived', sorted in descending order
print(corr_matrix['survived'].sort_values(ascending=False))

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
from sklearn.preprocessing import LabelEncoder

# Load Titanic dataset from seaborn
titanic_df = sns.load_dataset('titanic')

# 1. Label Encoding for 'sex' column
le = LabelEncoder()
titanic_df['sex_encoded'] = le.fit_transform(titanic_df['sex'])

# 2. One-Hot Encoding for 'embarked' column
# Drop the first column to avoid multicollinearity (drop_first=True)
titanic_df_encoded = pd.get_dummies(titanic_df, columns=['embarked'], drop_first=True)

# Display the first few rows to verify the encoding
print(titanic_df_encoded[['sex', 'sex_encoded', 'embarked_Q', 'embarked_S']].head())
```
